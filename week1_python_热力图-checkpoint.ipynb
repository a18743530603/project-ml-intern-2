```python
#
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
titanic=sns.load_dataset("titanic")
#
#
data=titanic.select_dtypes(include=["int64","float64"])
corr=data.corr()
#
plt.figure(figsize=(8,6))
sns.heatmap(corr,# 
            annot=True,# 
            cmap="coolwarm",# 
            center=0,# 0
            fmt=".2f" #  
                )
plt.title("Titanic Numerical Features Correlation Heatmap")
plt.show()
```
