```python
#平均数、中位数、相关系数、置信区间、方差、协方差
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
titanic=sns.load_dataset("titanic")
#相关性矩阵，只能计算数值型
#挑选出数值型列
data=titanic.select_dtypes(include=["int64","float64"])
corr=data.corr()
#可视化热力图
plt.figure(figsize=(8,6))
sns.heatmap(corr,# 相关性矩阵
            annot=True,# 显示数值
            cmap="coolwarm",# 红蓝渐变颜色
            center=0,# 颜色中点设为0，方便区分正负相关
            fmt=".2f" # 显示两位小数 
                )
plt.title("Titanic Numerical Features Correlation Heatmap")
plt.show()
```
