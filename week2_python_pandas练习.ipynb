```python
#做按舱位和性别分组的生存率透视表，并观察差异。
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
data=sns.load_dataset("titanic")
pivot=data.pivot_table(index="pclass",columns="sex",values="survived")
print(pivot)
```

```python
#可视化观察差异
pivot.plot(kind='bar')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Pclass and Sex')
plt.show()
```
